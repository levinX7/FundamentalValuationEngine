```python
import pandas as pd
import numpy as np

# --- 0) Start from your existing DFs ---
fc_raw = fundamental_clean.copy()
ts_raw = time_series_aligned.copy()

# --- 1) Normalize names used across tables (just _B -> _b) ---
rename_map = {
    'dcf_market_cap_B': 'dcf_market_cap_b',
    'ddm_market_cap_B': 'ddm_market_cap_b',
    'multiples_market_cap_B': 'multiples_market_cap_b',
    'median_market_cap_B': 'median_market_cap_b',
    'current_market_cap_B': 'current_market_cap_b',
    'forecast_market_cap_B': 'forecast_market_cap_b',
}
fc = fc_raw.rename(columns=rename_map)
ts = ts_raw.rename(columns=rename_map)

# Ticker hygiene
for df in (fc, ts):
    df['ticker'] = df['ticker'].astype(str).str.upper().str.strip()

# --- 2) BASIC_INFO ---
basic_info = (
    fc[['ticker', 'current_market_cap_b', 'current_price', 'sector']]
    .drop_duplicates('ticker')
    .sort_values('ticker')
    .reset_index(drop=True)
)

# ensure types
for col in ['current_market_cap_b', 'current_price']:
    basic_info[col] = pd.to_numeric(basic_info[col], errors='coerce')

# --- 3) FUNDAMENTAL + ID ---
fundamental = (
    fc[['ticker', 'multiples_market_cap_b', 'ddm_market_cap_b',
        'dcf_market_cap_b', 'median_market_cap_b', 'buy?']]
    .drop_duplicates('ticker')
    .sort_values('ticker')
    .reset_index(drop=True)
)

# numeric cols
for col in ['multiples_market_cap_b', 'ddm_market_cap_b', 'dcf_market_cap_b', 'median_market_cap_b']:
    fundamental[col] = pd.to_numeric(fundamental[col], errors='coerce')

# buy? -> boolean (handles yes/no/true/false/1/0, otherwise NA)
def _to_bool(x):
    if pd.isna(x): return pd.NA
    s = str(x).strip().lower()
    if s in {'yes','true','1','y','t'}: return True
    if s in {'no','false','0','n','f'}: return False
    return pd.NA

fundamental['buy?'] = fundamental['buy?'].apply(_to_bool).astype('boolean')

# add surrogate key
fundamental.insert(1, 'fundamental_id', range(1, len(fundamental) + 1))

# --- 4) FORECAST + ID ---
forecast = (
    fc[['ticker', 'forecast_price_30d', 'forecast_market_cap_b', 'forecast_change_%']]
    .drop_duplicates('ticker')
    .sort_values('ticker')
    .reset_index(drop=True)
)

for col in ['forecast_price_30d', 'forecast_market_cap_b', 'forecast_change_%']:
    forecast[col] = pd.to_numeric(forecast[col], errors='coerce')

forecast.insert(1, 'forecast_id', range(1, len(forecast) + 1))

# --- 5) RISK + ID (aligned to basic_info tickers) ---
risk = (
    ts[['ticker', 'cumulative_return', 'annualized_return', 'annualized_volatility', 'sharpe_ratio']]
    .drop_duplicates('ticker')
)
# keep only tickers we have in basic_info to maintain 1:1
risk = risk[risk['ticker'].isin(basic_info['ticker'])].sort_values('ticker').reset_index(drop=True)

for col in ['cumulative_return', 'annualized_return', 'annualized_volatility', 'sharpe_ratio']:
    risk[col] = pd.to_numeric(risk[col], errors='coerce')

risk.insert(1, 'risk_id', range(1, len(risk) + 1))
```
