```python
# Top 10 “buy” with the largest value gap (median mkt cap − current mkt cap)

# 1) Value gap among BUY tickers
buy_fund = fundamental_df.query("`buy?` == 'Yes'").copy()

# ensure numeric (in case of stray strings)
for col in ["median_market_cap_B", "current_market_cap_B", "forecast_market_cap_B",
            "current_price", "forecast_price_30d", "forecast_change_%"]:
    buy_fund[col] = pd.to_numeric(buy_fund[col], errors="coerce")

buy_fund["value_gap_B"] = buy_fund["median_market_cap_B"] - buy_fund["current_market_cap_B"]

top_value_gap = (
    buy_fund
    .dropna(subset=["value_gap_B"])
    .sort_values("value_gap_B", ascending=False)
    .loc[:, ["ticker","sector","current_market_cap_B","median_market_cap_B","value_gap_B"]]
    .head(10)
)

top_value_gap
```

```python
# Top 10 sectors by count among “buy”

top_buy_sectors = (
    buy_fund["sector"]
    .value_counts(dropna=False)
    .head(10)
    .rename_axis("sector")
    .reset_index(name="buy_count")
)

top_buy_sectors
```

```python
# Top 10 BUY on 30d forecast % change

top_30d_increase = (
    buy_fund
    .dropna(subset=["forecast_change_%"])
    .sort_values("forecast_change_%", ascending=False)
    .loc[:, ["ticker","sector","forecast_change_%","current_price","forecast_price_30d",
             "current_market_cap_B","forecast_market_cap_B"]]
    .head(10)
)

top_30d_increase
```

```python
# Highest volatility among BUY tickers (merge to bring volatility in)

vol_cols = ["ticker","annualized_volatility","sharpe_ratio","annualized_return","cumulative_return"]
buy_with_vol = (
    buy_fund[["ticker","sector","current_price","forecast_price_30d","forecast_change_%"]]
    .merge(time_series_df[vol_cols], on="ticker", how="left")
)

top_volatility = (
    buy_with_vol
    .dropna(subset=["annualized_volatility"])
    .sort_values("annualized_volatility", ascending=False)
    .loc[:, ["ticker","sector","annualized_volatility","sharpe_ratio",
             "annualized_return","cumulative_return","forecast_change_%","current_price"]]
    .head(10)
)

top_volatility
```

```python
#  Composite ranking for BUY tickers: low risk + growth + undervaluation

mix_base = (
    buy_fund[["ticker","sector","current_price","forecast_price_30d","forecast_change_%",
              "current_market_cap_B","median_market_cap_B"]]
    .merge(time_series_df[["ticker","annualized_volatility","sharpe_ratio"]], on="ticker", how="left")
    .copy()
)

# compute value gap
mix_base["value_gap_B"] = mix_base["median_market_cap_B"] - mix_base["current_market_cap_B"]

# keep rows with all needed pieces
mix = mix_base.dropna(subset=["annualized_volatility","forecast_change_%","value_gap_B"]).copy()

# ranks: lower volatility is better (ascending), higher change% & gap are better (descending)
mix["rank_vol"]  = mix["annualized_volatility"].rank(method="min", ascending=True)
mix["rank_px"]   = mix["forecast_change_%"].rank(method="min", ascending=False)
mix["rank_gap"]  = mix["value_gap_B"].rank(method="min", ascending=False)

# simple average score (lower = better mix). tweak weights if you like.
mix["score"] = (mix["rank_vol"] + mix["rank_px"] + mix["rank_gap"]) / 3.0

perfect_mix_top10 = (
    mix
    .sort_values("score", ascending=True)
    .loc[:, ["ticker","sector","annualized_volatility","forecast_change_%",
             "value_gap_B","current_price","forecast_price_30d","sharpe_ratio"]]
    .head(10)
)

perfect_mix_top10
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Robust BUY flag (0/1) in fundamentals so it can be included numerically
def buy_to_num(s):
    s = s.astype(str).str.strip().str.lower()
    return s.map({
        "yes":1, "true":1, "1":1, "y":1, "t":1,
        "no":0,  "false":0, "0":0, "n":0, "f":0
    }).fillna(0).astype("int8")

fund = fundamental_df.copy()
if "buy?" in fund.columns:
    fund["BUY_num"] = buy_to_num(fund["buy?"])
elif "buy" in fund.columns:
    fund["BUY_num"] = buy_to_num(fund["buy"])
else:
    fund["BUY_num"] = 0  # fallback so code runs

# 2) Merge fundamentals + time series on 'ticker'
merged = fund.merge(time_series_df, on="ticker", how="left")

# 3) Keep numeric columns only; drop all-NaN & zero-variance columns
num = merged.select_dtypes(include="number").copy()
num = num.loc[:, num.notna().sum() > 1]
num = num.loc[:, num.std(ddof=0) > 0]

# 4) Compute Pearson correlation matrix
corr = num.corr(method="pearson")

# 5) Plot heat map
h, w = corr.shape
fig_w = min(20, 0.5 * w + 4)
fig_h = min(20, 0.5 * h + 4)

fig, ax = plt.subplots(figsize=(fig_w, fig_h))
im = ax.imshow(corr.values, vmin=-1, vmax=1, interpolation="nearest")

ax.set_xticks(np.arange(w))
ax.set_yticks(np.arange(h))
ax.set_xticklabels(corr.columns, rotation=90, ha="right")
ax.set_yticklabels(corr.index)

cbar = plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
cbar.set_label("Pearson correlation")

ax.set_title("Correlation Heat Map (Pearson)")

# 6) Add numeric annotations (guarded to avoid clutter if matrix is large)
if h <= 25 and w <= 25:
    vals = np.round(corr.values, 2)  # 2 decimal places
    for i in range(h):
        for j in range(w):
            ax.text(
                j, i, f"{vals[i, j]:.2f}",
                ha="center", va="center", fontsize=7,
                color=("white" if abs(corr.iat[i, j]) > 0.5 else "black")
            )

plt.tight_layout()
plt.show()
```

```python
# === Is BUY performance significantly higher than NON-BUY? (one-sided test) ===

import pandas as pd
import numpy as np
from scipy import stats

# 1) Build BUY flag (0/1) and merge with returns
def buy_to_num(s):
    s = s.astype(str).str.strip().str.lower()
    return s.map({"yes":1,"true":1,"1":1,"y":1,"t":1,
                  "no":0,"false":0,"0":0,"n":0,"f":0}).fillna(0).astype("int8")

fund = fundamental_df.copy()
if "buy?" in fund.columns:
    fund["BUY_num"] = buy_to_num(fund["buy?"])
elif "buy" in fund.columns:
    fund["BUY_num"] = buy_to_num(fund["buy"])
else:
    raise ValueError("Need a 'buy?' or 'buy' column for the BUY flag.")

merged = fund[["ticker","BUY_num"]].merge(
    time_series_df[["ticker","annualized_return","cumulative_return"]].copy(),
    on="ticker", how="left"
)

# 2) Pick the target (prefer realized annualized_return; else cumulative_return)
target = "annualized_return" if "annualized_return" in merged.columns else (
         "cumulative_return" if "cumulative_return" in merged.columns else None)
if target is None:
    raise ValueError("No return column found. Add 'annualized_return' or 'cumulative_return' to time_series_df.")

df = merged.dropna(subset=[target, "BUY_num"])

# 3) Descriptive means
buy = df.loc[df["BUY_num"] == 1, target].astype(float)
non = df.loc[df["BUY_num"] == 0, target].astype(float)
print(f"BUY mean = {buy.mean():.4f}  (n={buy.size})")
print(f"NON mean = {non.mean():.4f}  (n={non.size})")
print(f"Diff (BUY - NON) = {buy.mean() - non.mean():.4f}")

# 4) Significance tests (one-sided: BUY > NON)
t_stat, p_welch = stats.ttest_ind(buy, non, equal_var=False, alternative="greater")
u_stat, p_mw    = stats.mannwhitneyu(buy, non, alternative="greater")

alpha = 0.05
sig_welch = p_welch < alpha
sig_mw    = p_mw    < alpha

print("\nWelch t-test (one-sided BUY>NON): "
      f"t={t_stat:.3f}, p={p_welch:.4g} -> {'SIGNIFICANT' if sig_welch else 'not significant'}")
print("Mann–Whitney U (one-sided BUY>NON): "
      f"U={u_stat:.0f}, p={p_mw:.4g} -> {'SIGNIFICANT' if sig_mw else 'not significant'}")

print("\nConclusion:",
      "BUY outperforms significantly (at α=0.05)" if (sig_welch and sig_mw and buy.mean()>non.mean())
      else "No statistically significant outperformance at α=0.05")
```

```python
from scipy import stats

# testing whether returns and risk (annualized_volatility) move together.


ALPHA = 0.05  # significance threshold

# 1) Build BUY flag (0/1) and merge minimal columns
def buy_to_num(s):
    s = s.astype(str).str.strip().str.lower()
    return s.map({"yes":1,"true":1,"1":1,"y":1,"t":1,
                  "no":0,"false":0,"0":0,"n":0,"f":0}).fillna(0).astype("int8")

fund = fundamental_df.copy()
if "buy?" in fund.columns:
    fund["BUY_num"] = buy_to_num(fund["buy?"])
elif "buy" in fund.columns:
    fund["BUY_num"] = buy_to_num(fund["buy"])
else:
    fund["BUY_num"] = 0

cols_ts = ["ticker","annualized_return","cumulative_return","annualized_volatility"]
cols_ts = [c for c in cols_ts if c in time_series_df.columns]
df = fund[["ticker","BUY_num"]].merge(time_series_df[cols_ts], on="ticker", how="left")

# 2) Pick target (prefer realized annualized_return)
target = "annualized_return" if "annualized_return" in df.columns else (
         "cumulative_return" if "cumulative_return" in df.columns else None)
if target is None:
    raise ValueError("Need 'annualized_return' or 'cumulative_return' in time_series_df.")

risk = "annualized_volatility"
if risk not in df.columns:
    raise ValueError("Need 'annualized_volatility' in time_series_df.")

# 3) Helper: Pearson r + 95% CI + significance text
def pearson_report(frame, label, alpha=ALPHA):
    t = frame.dropna(subset=[target, risk]).copy()
    x = t[risk].astype(float).values
    y = t[target].astype(float).values
    n = x.size
    if n < 3:
        print(f"{label}: not enough data (n={n})."); 
        return {"label":label, "n":n, "r":np.nan, "p":np.nan, "sig":False, "dir":"NA"}

    r, p = stats.pearsonr(x, y)
    # Fisher z 95% CI
    z = np.arctanh(r)
    se = 1/np.sqrt(n-3)
    z_lo, z_hi = z - stats.norm.ppf(0.975)*se, z + stats.norm.ppf(0.975)*se
    lo, hi = np.tanh(z_lo), np.tanh(z_hi)

    sig = p < alpha
    direction = "positive" if r > 0 else "negative"
    verdict = f"SIGNIFICANT {direction}" if sig else "not significant"

    print(f"{label}: n={n} | r={r:.3f}, p={p:.4g}, 95% CI [{lo:.3f}, {hi:.3f}] -> {verdict}")
    return {"label":label, "n":n, "r":r, "p":p, "sig":sig, "dir":direction if sig else "NA"}

# 4) Reports
res_all = pearson_report(df, "Overall")
res_buy = pearson_report(df[df["BUY_num"]==1], "Within BUY")
res_non = pearson_report(df[df["BUY_num"]==0], "Within NON-BUY")

# 5) Summary line
def tag(res):
    if res["n"] < 3 or np.isnan(res["p"]):
        return f"{res['label']}: insufficient data"
    return f"{res['label']}: {'SIGNIFICANT ' + res['dir'] if res['sig'] else 'not significant'} (p={res['p']:.4g}, r={res['r']:.3f})"

print("\nSummary:")
print(tag(res_all))
print(tag(res_buy))
print(tag(res_non))
```
